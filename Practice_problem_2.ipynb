# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
import matplotlib.pyplot as plt
```

# 5.a Distribution of glucose level--->Univariate analysis using histogram

```python
plt.figure(figsize=(8,5))
plt.hist(diabetes_data['Glucose'],color='r',edgecolor='y')
plt.xlabel('Glucose level')
plt.ylabel('Frequency')
plt.title('Glucose level distribution')
plt.show()
```

# 5.b. Scatter plot of Glucose level vs BMI

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Glucose', y='BMI', data=diabetes_data, alpha=0.7)
plt.title('Scatter Plot of Glucose vs. BMI')
plt.xlabel('Glucose Level (mg/dL)')
plt.ylabel('BMI')
plt.show()
```

# 5.c. Boxplot of BMI by Diabetes Outcome

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Outcome', y='BMI', data=diabetes_data,color='y')
plt.title('Boxplot of BMI by Diabetes Outcome')
plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.show()
```
